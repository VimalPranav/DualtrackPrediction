```python
# 1-time
import os
import pandas as pd

base_dir = "D:/val_data"
frames_dir = os.path.join(base_dir, "frames")
transfs_dir = os.path.join(base_dir, "transfs")

rows = []

for folder in os.listdir(frames_dir):  # 050, 051, etc.
    frame_subdir = os.path.join(frames_dir, folder)
    trans_subdir = os.path.join(transfs_dir, folder)

    if not os.path.isdir(frame_subdir):
        continue

    for file in os.listdir(frame_subdir):
        if file.endswith(".h5"):
            frame_path = os.path.join(frame_subdir, file)
            trans_path = os.path.join(trans_subdir, file)

            frame_path = os.path.join(frame_subdir, file).replace("\\", "/")
            trans_path = os.path.join(trans_subdir, file).replace("\\", "/")

            # check if matching transform exists
            if not os.path.exists(trans_path):
                print(f"Missing transfs for {file}")
                continue

            sweep_id = f"{folder}_{file.replace('.h5','')}"

            rows.append({
                "sweep_id": sweep_id,
                "raw_tus_rec_frames_path": frame_path,
                "raw_tus_rec_tforms_path": trans_path,
                "split": "val"
            })

df = pd.DataFrame(rows)

# save CSV
df.to_csv("tusrec_metadata.csv", index=False)

print("CSV created with", len(df), "entries")
```

```python
import torch.nn.functional as F

# normalizing
images = ds[0]['images'].astype(np.float32) / 255

# convert to tensor
images = torch.tensor(images)        # (N, H, W)
images = images.unsqueeze(1)         # (N, 1, H, W)

images = F.interpolate(
    images,
    size=(256, 256),   # IMPORTANT
    mode='bilinear',
    align_corners=False
)

images = images.unsqueeze(0) 
```

```python
seq_len = 16
stride = 4
outputs = []

with torch.no_grad():

    for i in range(0, images.shape[1] - seq_len, stride):

        chunk = images[:, i:i+seq_len]

        out = model({
            "local_encoder_inputs": chunk,
            "global_encoder_inputs": chunk
        })
        print(out)

        # take LAST motion prediction only
        outputs.append(out[0, -1])

outputs = torch.stack(outputs)
pred_xyz = outputs.cpu().numpy()
```

```python
seq_len = 16
rel_poses = []

with torch.no_grad():
    for i in range(len(images[0]) - seq_len + 1):
        chunk = images[:, i:i+seq_len]   # (1, 16, 1, H, W)
        out = model({
            "local_encoder_inputs": chunk,
            "global_encoder_inputs": chunk
        })
        print(out)
        # Take only the LAST predicted delta — it has full 16-frame context
        rel_poses.append(out[0, -1].cpu().numpy())   # (6,)

pred = np.array(rel_poses)   # (N-15, 6)
```

```python
pred_xyz = pred
print(pred_xyz.shape)
print(np.mean(np.abs(pred_xyz[:,3:])))
```

```python
from scipy.spatial.transform import Rotation as R
T_global = np.eye(4)

trajectory = []
trajectory.append(T_global[:3,3].copy())

for p in pred_xyz:
    tx, ty, tz, rx, ry, rz = p
    
    # TEMPORARILY ignore predicted rotation
    R_mat = np.eye(3)   # identity — no rotation
    
    T_global[:3, 3] += T_global[:3, :3] @ np.array([tx, ty, tz])
    T_global[:3, :3] = T_global[:3, :3] @ R_mat
    trajectory.append(T_global[:3, 3].copy())

pred_traj = np.array(trajectory)
```

```python
from scipy.spatial.transform import Rotation as R
T_global = np.eye(4)

trajectory = []
trajectory.append(T_global[:3,3].copy())

scale = 0.3023 / 0.1577   # ≈ 1.917

for p in pred_xyz:
    tx, ty, tz, rx, ry, rz = p
    
    # Scale translation, suppress rotation
    tx *= scale
    ty *= scale  
    tz *= scale
    
    R_mat = R.from_rotvec([rx, ry, rz]).as_matrix()
    
    # Body frame composition
    T_rel = np.eye(4)
    T_rel[:3, :3] = R_mat
    T_rel[:3, 3] = [tx, ty, tz]
    T_global = T_global @ T_rel
    trajectory.append(T_global[:3, 3].copy())
    
pred_traj = np.array(trajectory)
```

```python
from scipy.spatial.transform import Rotation as R
T_global = np.eye(4)

trajectory = []
trajectory.append(T_global[:3,3].copy())

rot_scale = 0.00042 / abs(pred[:, 3:].mean())   # per-axis if needed

for p in pred_xyz:
    tx, ty, tz, rx, ry, rz = p
    
    # Clamp rotation to GT-plausible range
    max_rot = 0.005   # 5× GT max as ceiling
    rx = np.clip(rx, -max_rot, max_rot)
    ry = np.clip(ry, -max_rot, max_rot)
    rz = np.clip(rz, -max_rot, max_rot)
    
    R_mat = R.from_rotvec([rx, ry, rz]).as_matrix()
    T_rel = np.eye(4)
    T_rel[:3, :3] = R_mat
    T_rel[:3, 3] = [tx * scale, ty * scale, tz * scale]
    T_global = T_global @ T_rel
    trajectory.append(T_global[:3, 3].copy())
    
pred_traj = np.array(trajectory)
```

```python
gt = ds[0]['tracking']   # (N, 4, 4)
gt_xyz = gt[:, :3, 3]   # (N, 3)

gt_xyz = gt_xyz[:len(pred_traj)] 

# Align origins
gt_xyz = gt_xyz - gt_xyz[0]
pred_traj_aligned = pred_traj - pred_traj[0]

# Compute best-fit rigid alignment of the predicted trajectory to the ground truth.
def align_trajectories(src, dst, allow_scaling=False):
    src_mean = src.mean(axis=0)
    dst_mean = dst.mean(axis=0)
    src_centered = src - src_mean
    dst_centered = dst - dst_mean

    cov = dst_centered.T @ src_centered / src.shape[0]
    U, S, Vt = np.linalg.svd(cov)
    R = U @ Vt
    if np.linalg.det(R) < 0:
        Vt[-1, :] *= -1
        R = U @ Vt

    if allow_scaling:
        scale = np.trace(np.diag(S)) / np.sum(src_centered**2)
    else:
        scale = 1.0

    aligned = (src_centered @ R.T) * scale + dst_mean
    return aligned, R, scale, src_mean, dst_mean

pred_traj_aligned, _, scale, _, _ = align_trajectories(pred_traj, gt_xyz, allow_scaling=False)
print(f"Applied rigid alignment to predicted trajectory (scale={scale:.6f})")
```
